```python
# Run this to install ipykernel, matplotlib and pillow
# !pip install ipykernel matplotlib pillow

# Import necessary libraries
from PIL import Image # Pillow library for image processing
import matplotlib.pyplot as plt # Matplotlib for visualisations
import numpy as np # view arrays
```

```python
# Load the image

#### SELECT YOUR IMAGE ####

image_path = 'one_by_two_pixels.png'
# image_path = 'two_by_three_pixels.png'

image = Image.open(image_path)
```

```python
# Display the image
plt.imshow(image)
plt.axis('off')  # Hide axis
plt.title("Loaded PNG Image")
plt.show()
```

```python
# Show the width and height of the image

width, height = image.size

print(f"Image width: {width} pixels")
print(f"Image height: {height} pixels")
```

```python
# See the RGB values for the image
# (R, G, B) values for each pixel

# convert the image to RGB format
image_rgb = image.convert("RGB")

# Get the pixel values
pixel_values = list(image_rgb.getdata())
print(f"pixel_values: {pixel_values}")
```

```python
# Separate out RGB channels into their own arrays/matrices

# convert image to array using numpy
image_array = np.array(image_rgb)

# Print the size of the array
print(f"""height, width, depth/channels: {image_array.shape}

""")

# Print the array
print(f"""image_array:
      
{image_array}

""")
```

```python
# Split the array into RGB channels
r_channel = image_array[:, :, 0]  # Red channel is the first depth/channel
g_channel = image_array[:, :, 1]  # Green channel is the second depth/channel
b_channel = image_array[:, :, 2]  # Blue channel is the third depth/channel

# Show the channel values
print(f"""
r_channel:
{r_channel}

g_channel:
{g_channel}

b_channel:
{b_channel}
""")
```
